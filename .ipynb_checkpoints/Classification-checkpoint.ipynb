## Description des differentes colonnes 
Transaction_ID : identifiant unique pour chaque transaction.

Transaction_Amount : la valeur monétaire de la transaction.

Transaction_Volume : la quantité ou le nombre d'éléments/actions impliqués dans la transaction.

Average_Transaction_Amount : montant moyen historique des transactions pour le compte.

Frequency_of_Transactions : fréquence à laquelle les transactions sont généralement effectuées par le compte.

Time_Since_Last_Transaction : temps écoulé depuis la dernière transaction.

Day_of_Week : jour de la semaine où la transaction a eu lieu.

Time_of_Day : L'heure de la journée à laquelle la transaction a eu lieu.

Âge : Âge du titulaire du compte.

Sexe : Sexe du titulaire du compte.

Revenu : Revenu du titulaire du compte.

Account_Type : Type de compte (par exemple, personnel, professionnel).
